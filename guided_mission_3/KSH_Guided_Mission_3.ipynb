위 결과에서 gtf 파일이 없다고 나오면 압축이 풀리지 않은 상태일 수 있으니, 미리 풀어서 작업해보세요.

Splicing 되지 않는 transcript이기 때문에 결과가 단조롭습니다. 해당 좌표 범위로 bam을 솎아냅니다.

````
import pandas as pd
import re
import math
from collections import Counter
````

````
def shannon_entropy(snp_string):
    frequency = Counter(snp_string)
    total_length = len(snp_string)
    probabilities = [freq / total_length for freq in frequency.values()]
    entropy = -sum(p * math.log2(p) for p in probabilities if p > 0)
    if entropy == -0.0:
      entropy = 0.0
    return entropy
````

````
def make_bedgraph(bamfile):
  # Make gene bam file
  gene_range = !grep -i "$bamfile" ../binfo1-datapack1/gencode.gtf | awk -F'\t' '$3=="gene"' | cut -f1,4,5
  chm, start, end = gene_range[0].split('\t')
  !samtools view -b -o CLIP-"$bamfile".bam ../binfo1-datapack1/CLIP-35L33G.bam "$chm":"$start"-"$end"

  # Make pileup file
  all_bamfile = f"CLIP-{bamfile}.bam"
  all_pileupfile = f"CLIP-{bamfile}.pileup"
  gene_pileupfile = f"CLIP-{bamfile}-gene.pileup"
  !samtools mpileup "$all_bamfile" > "$all_pileupfile"

  # Filter pileup file
  !awk -v start="$start" -v end="$end" '$2 >= start && $2 <= end' "$all_pileupfile" > "$gene_pileupfile"

  # Process pileup file
  pileup = pd.read_csv(gene_pileupfile, sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
  toremove = re.compile('[<>$*#^]')
  pileup['matches'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))
  my_data = pileup[['chrom', 'pos', 'matches']].copy()
  my_data['entropy'] = my_data['matches'].apply(shannon_entropy)

  # Make bedgraph
  bedgraph_lines = []
  for index, row in my_data.iterrows():
      chrom = row['chrom']
      start = row['pos']
      end = start + 1
      value = row['entropy']
      bedgraph_lines.append(f"{chrom}\t{start}\t{end}\t{value}")

  # Save bedgraph
  bedfile = f"CLIP-{bamfile}-bedgraph.txt"
  with open(bedfile, 'w') as f:
    for line in bedgraph_lines:
        f.write(line + '\n')


````

````
make_bedgraph("mirlet7g")
make_bedgraph("mirlet7f-1")
make_bedgraph("mirlet7d")
````
